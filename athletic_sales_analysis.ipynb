```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020 = "Resources/athletic_sales_2020.csv"
sales_2021 = "Resources/athletic_sales_2021.csv"
sales_2020_df = pd.read_csv(sales_2020)
sales_2021_df = pd.read_csv(sales_2021)
```

```python
# Display the 2020 sales DataFrame
sales_2020_df
```

```python
# Display the 2021 sales DataFrame
sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.info()
```

```python
# Check the 2021 sales data types.
sales_2021_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
merged_df = pd.concat([sales_2020_df, sales_2021_df], ignore_index=True)
merged_df.reset_index(drop=True, inplace=True)
merged_df
```

```python
# Check if any values are null.
merged_df.isnull()
```

```python
# Check the data type of each column
merged_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
merged_df['invoice_date'] = pd.to_datetime(merged_df['invoice_date'])
merged_df
```

```python
# Confirm that the "invoice_date" data type has been changed.
merged_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
region_df = merged_df.groupby(['region','state','city'])['units_sold'].sum().reset_index()
region_df.rename(columns={'units_sold':'total_products_sold'},inplace=True)
#sort by total_products_sold
region_df = region_df.sort_values(by='total_products_sold', ascending=False)
# Show the top 5 results.
region_df.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table = region_df.pivot_table(
    index=['region', 'state', 'city'],
    values='total_products_sold',
    aggfunc='sum'
).reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
# already done in the previous block

# Show the top 5 results.
pivot_table.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
region_sales = merged_df.groupby(['retailer','region','state','city',])['total_sales'].sum().reset_index()
region_sales.rename(columns={'total_sales':'Total_Sales'},inplace=True)
region_sales = region_sales.sort_values(by='Total_Sales', ascending=False)
# Show the top 5 results.
region_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_top_products = pd.pivot_table(merged_df,
                                    values='total_sales',
                                    index=['region', 'state', 'city'],
                                    aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_top_products.rename(columns={'total_sales':'Total_Sales'},inplace=True)
pivot_top_products = pivot_top_products.sort_values(by='Total_Sales', ascending=False)
# Show the top 5 results.
pivot_top_products.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear = merged_df[merged_df['product'] == "Women's Athletic Footwear"]
womens_athletic_footwear.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
groupby_womens_athletic_footwear = womens_athletic_footwear.groupby(['retailer','region','state','city',])['total_sales'].sum().reset_index()
groupby_womens_athletic_footwear = groupby_womens_athletic_footwear.sort_values(by='total_sales', ascending=False)
# Show the top 5 results.
groupby_womens_athletic_footwear.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_womens_athletic_footwear = pd.pivot_table(womens_athletic_footwear, 
                             values='units_sold',
                             index = ['retailer','region', 'state', 'city'], 
                             aggfunc = 'sum').reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_womens_athletic_footwear.rename(columns= {'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
pivot_womens_athletic_footwear = pivot_womens_athletic_footwear.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)

# Show the top 5 results.
pivot_womens_athletic_footwear.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
invoice_and_sales = pd.pivot_table(womens_athletic_footwear,
                                 values='total_sales',
                                 index='invoice_date',
                                 aggfunc='sum')
invoice_and_sales.sort_values(by='total_sales', ascending=False, inplace=True)
# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
invoice_and_sales
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
bins = [0, 5, 10, 15, 20, 25 ,31]
group_labels = ['0 to 5','6 to 10', '11 to 15', '16 to 20', '21 to 25', '26 to 31']

# Sort the resampled pivot table in descending order on "Total Sales".
invoice_and_sales['total_sales'] = pd.cut(invoice_and_sales['total_sales'], bins, labels=group_labels)
invoice_and_sales.head()
```

```python
# Resample the data into daily bins and sum the sales for each day
daily_sales = invoice_and_sales.resample('D').sum()

# Sort the resampled data in descending order of 'Total Sales'
sorted_daily_sales = daily_sales.sort_values(by='total sales', ascending=False)

sorted_daily_sales
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = invoice_and_sales.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
sorted_weekly_sales = weekly_sales.sort_values(by='total sales', ascending=False)
```
